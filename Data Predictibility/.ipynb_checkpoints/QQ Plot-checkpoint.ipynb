```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats
sns.set()
import pylab
```

```python
df = pd.read_csv('abc.csv',parse_dates=['date'],index_col='date')
df.head()
```

```python
plt.rcParams['figure.figsize'] = 25, 20
fig, ax = plt.subplots(4, 2)

# no2
scipy.stats.probplot(df.no2, plot = ax[0, 0])
ax[0, 0].set_title('QQ Plot for NO2', size=24)

sns.distplot(df.no2, hist=False, ax = ax[0, 1])
ax[0, 1].set_title('Density Plot for NO2', size=24)


# so2
scipy.stats.probplot(df.so2, plot = ax[1, 0])
ax[1, 0].set_title('QQ Plot for SO2', size=24)

sns.distplot(df.so2, hist=False, ax = ax[1, 1])
ax[1, 1].set_title('Density Plot for SO2', size=24)


# rspm
scipy.stats.probplot(df.rspm, plot = ax[2, 0])
ax[2, 0].set_title('QQ Plot for RSPM', size=24)

sns.distplot(df.rspm, hist=False, ax = ax[2, 1])
ax[2, 1].set_title('Density Plot for RSPM', size=24)


# spm
scipy.stats.probplot(df.spm, plot = ax[3, 0])
ax[3, 0].set_title('QQ Plot for SPM', size=24)

sns.distplot(df.spm, hist=False, ax = ax[3, 1])
ax[3, 1].set_title('Density Plot for NIKKEI', size=24)

fig.tight_layout(pad=3.0)
pylab.show()
```
